# Use Pandas to plot the $CO_2$ data from the previous exercise

```python
%matplotlib inline   

import urllib2
import StringIO 
import itertools 
import datetime
import numpy as np
import matplotlib.pyplot as plt
import string
import pandas as pd
```

```python


url = "ftp://aftp.cmdl.noaa.gov/products/trends/co2/co2_mm_gl.txt"


df = pd.read_csv(url, sep=' +', index_col=0, skiprows=58, 
                 na_values='-998', engine='python',
                 header=None,
                 names=['year', 'month', 'decimal', 'average', 'trend'],)

fig = plt.figure(1,dpi=300,figsize=(16,12)) 

ax1 = plt.subplot(2,1,1)

sdf = df.set_index(['decimal'], inplace=True)
ax1.plot(df['average'],'rs-')
ax1.plot(df['trend'], '.-')

plt.title('Recent Global CO$_2$') 
plt.xlabel('Year')
plt.ylabel(u'CO$_2$ Concentration [PPM]')
ax1.legend(loc='best',fontsize=12,shadow=True)
```
